# Importing the Libraries

```python
import numpy as np
import pandas as pd
import sklearn
```

# Now we will load the data that we will be using to train our model

```python
from sklearn.datasets import load_boston
df = load_boston()
```

```python
df.keys()  # Returns all the keys of the dataset dictionary
```

```python
print(df.data)  #Info about the dataset
```

# We convert our dataset into the pandas dataframe , so that it is easier to analyse the data

```python
boston = pd.DataFrame(df.data, columns=df.feature_names)
boston.head(10)
```

```python
boston['MEDV'] = df.target
boston.head()
```

# Cheack if the dataset contains any null value or not

```python
boston.isnull()
```

```python
boston.isnull().sum()
```

# We never train the model on all the data that we have , we always make sure to atleast have a test dataset,which is different from the training dataset.

```python
from sklearn.model_selection import train_test_split
X = boston.drop('MEDV', axis=1)
Y = boston['MEDV']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.15, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)


```

# Lets import the linear Regression model from sklearn and train  it on the training dataset

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
from sklearn import linear_model
# Fitting model on the traning dataset

lin_model = LinearRegression()

lin_model.fit(X_train, Y_train)
```

```python
y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train,y_train_predict)))

print("The model performance for trainig set")
print('RMSE is {}'.format(rmse))
print("\n")

# on testing set
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))

print("The model performance for testing set")
print('RMSE is {}'.format(rmse))
```

```python
# MATPLOATLIB is a library to ploat the curves
# part of data visualization

import matplotlib.pyplot as plt

plt.figure(figsize=(5, 5))
plt.scatter(Y_test, y_test_predict)
plt.ploat([min(y_test_predict),max(y_test_predict)],[min(y_test_predict)],max(y_test_predict))
plt.xlabel('Actual')
plt.ylabel('Predicted')
```
